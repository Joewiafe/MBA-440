```python
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Joewiafe/MBA-440/main/Chipotle.csv"
Alinco = pd.read_csv(url)
```

```python
Alinco.head(10)
#provides the first 10 raw of data. this provides a quick view of the dataset titled "tips"
```

```python
Alinco.describe()
#quickly provides a descriptive statistics for all the continous variables in the dataset
```

```python
Alinco.info()
#provides information about each variable in the dataset. Dtype of float is a number with a decimal place, while int is for integers
```

```python
Alinco['gender'].value_counts()
#"gender" is a categorical variable in our data
```

```python
Alinco['top1'].value_counts()
#"top1" is a categorical variable in our data
```

```python
Alinco.corr()
#quickly provides a correlation matrix
```

```python
sns.heatmap(Alinco.iloc[:,7:12])
#annot=True adds an annotation to the right

```

```python
sns.heatmap(Alinco.corr(), cmap="YlGnBu")
#just changing colors here for better looks (darker=greater value)
#other colormap options are 'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds','YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu','GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'Bu
```

```python
sns.scatterplot(x="age", y="income", hue="gender", data=Alinco)
#adding color-coding by male-female by adding the hue statement
```

```python
sns.lmplot(x = "age", y = "income", hue="gender", data = Alinco)
#Fitting lines for the gender categories
```

```python
sns.countplot(x='gender', data=Alinco, color='r')
#changing color to all red by adding "color='r'"
```

```python
Alinco_gender=Alinco.groupby("gender")
Alinco_gender.mean()
#groupby() is used to split the data in dataframe into groups ("Alinco_gender" is created as a new dataframe)
#"gender" is a categorical variable
```

```python
sns.boxplot(x="gender", y="age", data=Alinco)
#quickly creating boxplots. Here, "gender" is a categorical variable" and "age" is a continuous variable
```
